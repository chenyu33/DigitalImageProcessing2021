## canny edge detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('sIMG_8253.JPG',0)
edges = cv2.Canny(img,100,200)
blur = cv2.GaussianBlur(img,(3,3),0)
edges1 = cv2.Canny(blur,100,120)
plt.figure(figsize=(10,10))
plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image with smooth'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
im = cv2.imread('sIMG_8253.JPG')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(im,100,200)
blur = cv2.GaussianBlur(im,(3,3),0)
edges1 = cv2.Canny(blur,100,120)
ret,thresh = cv2.threshold(edges1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.drawContours(im, contours, -1, (255,0,0), 3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(im,cmap = 'gray')
plt.title('im'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.show()
```

## contour analysis

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('sIMG_8255.JPG')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(im,(9,9),0)
blur2 = cv2.GaussianBlur(im,(9,9),0)
imgray2 = cv2.cvtColor(blur2,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.drawContours(im, contours, -1, (255,0,0), 3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(131),plt.imshow(imgray,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(im,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.show()
```
